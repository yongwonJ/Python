```python
import cv2

#low_threshold
#high_threshold
#Canny(image, low, high) documentation 찾아보기
#가우시안 블러? 원리 보기


print(cv2.__version__)
cap = cv2.VideoCapture(0)#연결된 카메라를 연결 - 카메라 준비
cap.set(3, 200) #x
cap.set(4, 400) #
low = 50
high = 150


while True:
    ret, frame = cap.read()
    #_, frame = cap.read()
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)#cv에선 bgr - 옛날엔 bgr이 쉬워서 이렇게 했었다.
    # rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2LAB)#cv에선 bgr - 옛날엔 bgr이 쉬워서 이렇게 했었다.
    canny_frame = cv2.Canny(gray_frame, low, high)
    blur_frame = cv2.GaussianBlur(frame, (3,3), 0)

    print(ret)

    cv2.imshow("Video", frame)
    cv2.imshow('Gray', gray_frame)
    #cv2.imshow('RGB', rgb_frame)
    cv2.imshow('Canny', canny_frame)
    cv2.imshow("Blur", blur_frame)
    if cv2.waitKey(10) & 0xff == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
